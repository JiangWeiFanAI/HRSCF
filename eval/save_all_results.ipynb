```python
# fro HRFSR_v1

from datetime import timedelta, date, datetime
import os
import sys
sys.path.append("../")
import util.data_processing_tool as dpt
import numpy as np
import csv
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.backends.backend_pdf import PdfPages
import matplotlib
matplotlib.use('Agg')


total_leading=217
interval_band=[90,75,50,25,10]
accumulate_days=[7,14,30,45,60,217]


class aaa(object):
    def __init__(self,lead,year_generate=1997):
        self.ensemble_access=['e01','e02','e03','e04','e05','e06','e07','e08','e09','e10','e11']
        self.lead_time=lead
        self.year_generate=year_generate
        self.files=self.get_filename_with_time_order()
    def get_filename_with_time_order(self):
        _files = []
        for mm in range(1,13):
            for dd in [1,9,17,25]:
#                 for i in range(self.lead_time,self.lead_time+1):
#                 for en in ['e01','e02','e03','e04','e05','e06','e07','e08','e09','e10','e11']:
                path=[]
                date_time=date(self.year_generate, mm, dd)
                barra_date=date_time+timedelta(self.lead_time)
                path.append(date_time)
                path.append(barra_date)
                path.append(self.lead_time)
                _files.append(path)
        return _files
    def __getitem__(self,idx):
        return self.files[idx]

def load_climatology_data(l,measurement ,year_generate,climatology_num):
    data=aaa(l,year_generate)
    climtology_lead_time=[]
    climatology_data=np.load('../save/'+measurement+'/climatology/climatology_'+str(year_generate)+'_varify_lead_time_windows_'+str(climatology_num)+'.npy')
#     print(climatology_data.shape)
    dates_needs=date_range(date(year_generate, 1, 1),date(year_generate+1, 7, 29))
    date_map=np.array(dates_needs)
    for _,target_date,_ in data.files:
        idx=np.where(date_map==target_date)[0]
        climtology_lead_time.append(climatology_data[idx][0])
    return np.array(climtology_lead_time)

def date_range(start_date, end_date):
    """This function takes a start date and an end date as datetime date objects.
    It returns a list of dates for each date in order starting at the first date and ending with the last date"""
    return [start_date + timedelta(x) for x in range((end_date - start_date).days + 1)]







def table_csv_confidence(data_np,file_dir):
    with open(file_dir+'.csv', "w", newline='') as file:
        csv_file = csv.writer(file)
        head = ["leading",str(interval_band[0]),str(interval_band[1]),str(interval_band[2]),str(interval_band[3]),str(interval_band[4])]
        csv_file.writerow(head)

        for lead_time in range(total_leading):
            line=[lead_time,data_np[0,lead_time],data_np[1,lead_time],data_np[2,lead_time],data_np[3,lead_time],data_np[4,lead_time]]

            csv_file.writerow(line)

def draw_plot(data,name,ylim_set=(-0.6,0.5),param='CRPS_SS'):
    if param=='CRPS_SS':
        ylim_set=(-0.6,0.5)
    else:
        ylim_set=(-0.6,1)
        
    x=list(range(1,total_leading+1))
    data_np=np.array(data).T
    # plt.plot(x,mean_my,label="",color="b")#mean
    plt.figure(dpi=100,figsize=(15,5))
    plt.plot(x,data_np[2,:],label="",color="r")#median
    plt.plot(x,[0]*total_leading,color="#000000")
    plt.fill_between(x,data_np[0,:],data_np[1,:],color="#cacaca")
    plt.fill_between(x,data_np[1,:],data_np[2,:],color="#989898")
    plt.fill_between(x,data_np[2,:],data_np[3,:],color="#989898")
    plt.fill_between(x,data_np[3,:],data_np[4,:],color="#cacaca")
    # plt.xlim(0,217)
    # plt.grid()
    plt.xlabel(" Leadtime (day)")
    plt.ylabel(param)
    plt.ylim(ylim_set[0],ylim_set[1])
    pdf = PdfPages(name+".pdf")
    pdf.savefig()
    pdf.close()
    plt.clf()
    table_csv_confidence(data_np,name)
            


            
            
def save_csv_mean(my_pr=0,my_pr_zg=0,cali=0,bcsd=0,BI=0,climat=0,ss_my_pr=0,ss_my_pr_zg=0,ss_cali=0,ss_bcsd=0,ss_BI=0,lead=0,file_dir=0,first_init=0,climatology_num=1):
    if first_init:
        with open(file_dir+'_mean.csv', "w", newline='') as file:
            csv_file = csv.writer(file)
            head = ["leading","climatology_"+str(climatology_num),'bilinear interpolation','vdsrd','vdsrd2','calibration','bcsd','crps_ss_mean_bilinearinterpolation_against_climatology','crps_ss_mean_vdsrd_against_climatology','crps_ss_mean_vdsrd2_against_climatology','crps_ss_mean_calibration_against_climatology','crps_ss_mean_bcsd_against_climatology']
            csv_file.writerow(head)
    else:
        line=[lead,climat.mean(),BI.mean(),my_pr.mean(),my_pr_zg.mean(),cali.mean(),bcsd.mean(),ss_BI.mean(), ss_my_pr.mean(),ss_my_pr_zg.mean(),ss_cali.mean(),ss_bcsd.mean()]
        with open(file_dir+'_mean.csv', "a", newline='') as file:
            csv_file = csv.writer(file)
#             head = ["leading","climatology_"+str(climatology_num),'bilinear interpolation','vdsrd','vdsrd2','calibration','cspr_ss_mean_bilinearinterpolation_against_climatology','cspr_ss_mean_vdsrd_against_climatology','cspr_ss_mean_vdsrd2_against_climatology','cspr_ss_mean_calibration_against_climatology']
            csv_file.writerow(line) 
    
def save_csv_median(my_pr=0,my_pr_zg=0,cali=0,bcsd=0,BI=0,climat=0,ss_my_pr=0,ss_my_pr_zg=0,ss_cali=0,ss_bcsd=0,ss_BI=0,lead=0,file_dir=0,first_init=0,climatology_num=1):
    if first_init:
        with open(file_dir+'_median.csv', "w", newline='') as file:
            csv_file = csv.writer(file)
            head = ["leading","climatology_"+str(climatology_num),'bilinear interpolation','vdsrd','vdsrd2','calibration','bcsd','crps_ss_mean_bilinearinterpolation_against_climatology','crps_ss_mean_vdsrd_against_climatology','crps_ss_mean_vdsrd2_against_climatology','crps_ss_mean_calibration_against_climatology','crps_ss_mean_bcsd_against_climatology']
            csv_file.writerow(head)
    else:
        line=[lead,np.median(climat),np.median(BI),np.median(my_pr),np.median(my_pr_zg),np.median(cali),np.median(bcsd).,np.median(ss_BI), np.median(ss_my_pr),np.median(ss_my_pr_zg),np.median(ss_cali),np.median(ss_bcsd)]
        with open(file_dir+'_median.csv', "a", newline='') as file:
            csv_file = csv.writer(file)
#             head = ["leading","climatology_"+str(climatology_num),'bilinear interpolation','vdsrd','vdsrd2','calibration','cspr_ss_mean_bilinearinterpolation_against_climatology','cspr_ss_mean_vdsrd_against_climatology','cspr_ss_mean_vdsrd2_against_climatology','cspr_ss_mean_calibration_against_climatology']
            csv_file.writerow(line)           
            
            
            
def draw_confidence_figure(measurement='crps',climatology_num=1,option='whole',year_generate=2010):
    for i in ['bi','vdsrd','vdsrd2','calibration','bcsd']:
        for j in ['50_station','whole']:
            dpt.mkdir('../save/results/'+measurement+'/'+str(year_generate)+'/accumulated_leading_time_figures/'+i+'/'+j)
            dpt.mkdir('../save/results/'+measurement+'/'+str(year_generate)+'/single_leading_time_figures/'+i+'/'+j)
            for e in ['skill_score_with_calimatology_1','skill_score_with_calimatology_11']:
                dpt.mkdir('../save/results/'+measurement+'/'+str(year_generate)+'/accumulated_leading_time_figures/'+i+'/'+e+'/'+j)
                dpt.mkdir('../save/results/'+measurement+'/'+str(year_generate)+'/single_leading_time_figures/'+i+'/'+e+'/'+j)

    
    if option == "whole":
        save_path='../save/results/'+measurement+'/'+str(year_generate)

        file_path=save_path+'/table_'+option+'_' +str(year_generate)+'_climatology_' +str(climatology_num)
        
        
        save_csv_mean(file_dir=file_path,first_init=1)
        save_csv_median(file_dir=file_path,first_init=1)

        land=np.load("../save/mae/calibration/1997/lead_time_"+str(216)+".npy").mean(0)




        data_my_pr_against_climatology=[]
        data_my_pr_zg_against_climatology=[]
        data_cali_against_climatology=[]
        data_BI_against_climatology=[]
        data_bcsd_against_climatology=[]

        
        mean_my=[]


            

        acc_pr=np.zeros(land.shape)
        acc_pr_zg=np.zeros(land.shape)
        acc_bi=np.zeros(land.shape)
        acc_cali=np.zeros(land.shape)
        acc_bcsd=np.zeros(land.shape)

        acc_ss_pr=np.zeros(land.shape)
        acc_ss_pr_zg=np.zeros(land.shape)
        acc_ss_bi=np.zeros(land.shape)
        acc_ss_cali=np.zeros(land.shape)
        acc_ss_bcsd=np.zeros(land.shape)

        for lead_time in range(total_leading):
            my_pr=np.load("../save/"+measurement+"/vdsrd/"+str(year_generate)+"/lead_time_"+str(lead_time)+".npy").mean(0)
            my_pr_zg=np.load("../save/"+measurement+"/vdsrd2/"+str(year_generate)+"/lead_time_"+str(lead_time)+".npy").mean(0)
            BI=np.load("../save/"+measurement+"/bi_217/"+str(year_generate)+"/lead_time_"+str(lead_time)+".npy").mean(0)
            climat=load_climatology_data(lead_time,measurement,year_generate,climatology_num).mean(0)
            cali=np.load("../save/"+measurement+"/calibration/"+str(year_generate)+"/lead_time_"+str(lead_time)+".npy").mean(0)
            bcsd=np.load("../save/"+measurement+"/bcsd/"+str(year_generate)+"/lead_time_"+str(lead_time)+".npy").mean(0)
            
            acc_pr+=my_pr
            acc_pr_zg+=my_pr_zg
            acc_bi+=BI
            acc_cali+=cali
            acc_bcsd+=bcsd


#               #####################draw overview figure####################  
            my_pr,lat,lon=dpt.add_lat_lon_data(my_pr)
            # # draw_aus_figure(my_pr,lat,lon,param='crps_ss')
            # draw_station_figure(my_pr,param='mae',title='')

            if measurement=='crps':
                ss_my_pr= 1-my_pr/climat
                ss_my_pr_zg=1-my_pr_zg/climat
                ss_cali=1-cali/climat
                ss_bi=1-BI/climat
                ss_bcsd=1-bcsd/climat


#                     acc_ss_pr= 1-(acc_pr/climat)/(lead_time+1)
#                     acc_ss_pr_zg=1-acc_pr_zg/(climat* (lead_time+1) )
#                     acc_ss_cali=1-acc_cali/(climat*((lead_time+1)))
#                     acc_ss_bi=1-acc_bi/(climat*((lead_time+1)))                    

#                     acc_ss_pr+=ss_my_pr
#                     acc_ss_pr_zg+=ss_my_pr_zg
#                     acc_ss_bi+=ss_bi
#                     acc_ss_cali+=ss_cali

            if lead_time+1 in accumulate_days:

                if climatology_num==11:  # draw once
                    dpt.draw_aus_figure(BI,lat,lon,param=measurement,title=measurement.upper(),cmap_label='',save=True,path=save_path+'/'+'single_leading_time_figures/bi/whole/bi_whole_single_leading_time_'+str(lead_time+1))
                    dpt.draw_aus_figure(my_pr_zg,lat,lon,param=measurement,title=measurement.upper(),cmap_label='',save=True,path=save_path+'/'+'single_leading_time_figures/vdsrd2/whole/vdsrd2_whole_single_leading_time_'+str(lead_time+1))
                    dpt.draw_aus_figure(my_pr,lat,lon,param=measurement,title=measurement.upper(),cmap_label='',save=True,path=save_path+'/'+'single_leading_time_figures/vdsrd/whole/vdsrd_whole_single_leading_time_'+str(lead_time+1))
                    dpt.draw_aus_figure(cali,lat,lon,param=measurement,title=measurement.upper(),cmap_label='',save=True,path=save_path+'/'+'single_leading_time_figures/calibration/whole/calibration_whole_single_leading_time_'+str(lead_time+1))
                    dpt.draw_aus_figure(climat,lat,lon,param=measurement,title=measurement.upper(),cmap_label='',save=True,path=save_path+'/'+'single_leading_time_figures/climatology_whole_'+str(climatology_num))
                    dpt.draw_station_figure(BI,param=measurement,title=measurement.upper(),cmap_label='',save=True,path=save_path+'/'+'single_leading_time_figures/bi/50_station/bi_50_station_single_leading_time_'+str(lead_time+1))
                    dpt.draw_station_figure(my_pr_zg,param=measurement,title=measurement.upper(),cmap_label='',save=True,path=save_path+'/'+'single_leading_time_figures/vdsrd2/50_station/vdsrd2_50_station_single_leading_time_'+str(lead_time+1))
                    dpt.draw_station_figure(my_pr,param=measurement,title=measurement.upper(),cmap_label='',save=True,path=save_path+'/'+'single_leading_time_figures/vdsrd/50_station/vdsrd_50_station_single_leading_time_'+str(lead_time+1))
                    dpt.draw_station_figure(cali,param=measurement,title=measurement.upper(),cmap_label='',save=True,path=save_path+'/'+'single_leading_time_figures/calibration/50_station/calibration_50_station_single_leading_time_'+str(lead_time+1))
                    dpt.draw_station_figure(climat,param=measurement,title=measurement.upper(),cmap_label='',save=True,path=save_path+'/'+'single_leading_time_figures/climatology_50_station_'+str(climatology_num))

                    dpt.draw_aus_figure(acc_bi/(lead_time+1),lat,lon,param=measurement,title=measurement.upper(),cmap_label='',save=True,path=save_path+'/'+'accumulated_leading_time_figures/bi/whole/bi_whole_accumulated_leading_time_'+str(lead_time+1))
                    dpt.draw_aus_figure(acc_pr_zg/(lead_time+1),lat,lon,param=measurement,title=measurement.upper(),cmap_label='',save=True,path=save_path+'/'+'accumulated_leading_time_figures/vdsrd2/whole/vdsrd2_whole_accumulated_leading_time_'+str(lead_time+1))
                    dpt.draw_aus_figure(acc_pr/(lead_time+1),lat,lon,param=measurement,title=measurement.upper(),cmap_label='',save=True,path=save_path+'/'+'accumulated_leading_time_figures/vdsrd/whole/vdsrd_whole_accumulated_leading_time_'+str(lead_time+1))
                    dpt.draw_aus_figure(acc_cali/(lead_time+1),lat,lon,param=measurement,title=measurement.upper(),cmap_label='',save=True,path=save_path+'/'+'accumulated_leading_time_figures/calibration/whole/calibration_whole_accumulated_leading_time_'+str(lead_time+1))
                    dpt.draw_station_figure(acc_bi/(lead_time+1),param=measurement,title=measurement.upper(),cmap_label='',save=True,path=save_path+'/'+'accumulated_leading_time_figures/bi/50_station/bi_50_station_accumulated_leading_time_'+str(lead_time+1))
                    dpt.draw_station_figure(acc_pr_zg/(lead_time+1),param=measurement,title=measurement.upper(),cmap_label='',save=True,path=save_path+'/'+'accumulated_leading_time_figures/vdsrd2/50_station/vdsrd2_50_station_accumulated_leading_time_'+str(lead_time+1))
                    dpt.draw_station_figure(acc_pr/(lead_time+1),param=measurement,title=measurement.upper(),cmap_label='',save=True,path=save_path+'/'+'accumulated_leading_time_figures/vdsrd/50_station/vdsrd_50_station_accumulated_leading_time_'+str(lead_time+1))
                    dpt.draw_station_figure(acc_cali/(lead_time+1),param=measurement,title=measurement.upper(),cmap_label='',save=True,path=save_path+'/'+'accumulated_leading_time_figures/calibration/50_station/calibration_50_station_accumulated_leading_time_'+str(lead_time+1))

                if measurement=='crps':
                    dpt.draw_aus_figure(ss_bi,lat,lon,param='crps_ss',title=measurement.upper()+'_SS',cmap_label='',save=True,path=save_path+'/'+'single_leading_time_figures/bi/skill_score_with_calimatology_'+str(climatology_num)+'/whole/bi_whole_single_leading_time_'+str(lead_time+1))
                    dpt.draw_aus_figure(ss_my_pr_zg,lat,lon,param='crps_ss',title=measurement.upper()+'_SS',cmap_label='',save=True,path=save_path+'/'+'single_leading_time_figures/vdsrd2/skill_score_with_calimatology_'+str(climatology_num)+'/whole/vdsrd2_whole_single_leading_time_'+str(lead_time+1))
                    dpt.draw_aus_figure(ss_my_pr,lat,lon,param='crps_ss',title=measurement.upper()+'_SS',cmap_label='',save=True,path=save_path+'/'+'single_leading_time_figures/vdsrd/skill_score_with_calimatology_'+str(climatology_num)+'/whole/vdsrd_whole_single_leading_time_'+str(lead_time+1))
                    dpt.draw_aus_figure(ss_cali,lat,lon,param='crps_ss',title=measurement.upper()+'_SS',cmap_label='',save=True,path=save_path+'/'+'single_leading_time_figures/calibration/skill_score_with_calimatology_'+str(climatology_num)+'/whole/calibration_whole_single_leading_time_'+str(lead_time+1))
                    dpt.draw_station_figure(ss_bi,param='crps_ss',title=measurement.upper()+'_SS',cmap_label='',save=True,path=save_path+'/'+'single_leading_time_figures/bi/skill_score_with_calimatology_'+str(climatology_num)+'/50_station/bi_50_station_single_leading_time_'+str(lead_time+1))
                    dpt.draw_station_figure(ss_my_pr_zg,param='crps_ss',title=measurement.upper()+'_SS',cmap_label='',save=True,path=save_path+'/'+'single_leading_time_figures/vdsrd2/skill_score_with_calimatology_'+str(climatology_num)+'/50_station/vdsrd2_50_station_single_leading_time_'+str(lead_time+1))
                    dpt.draw_station_figure(ss_my_pr,param='crps_ss',title=measurement.upper()+'_SS',cmap_label='',save=True,path=save_path+'/'+'single_leading_time_figures/vdsrd/skill_score_with_calimatology_'+str(climatology_num)+'/50_station/vdsrd_50_station_single_leading_time_'+str(lead_time+1))
                    dpt.draw_station_figure(ss_cali,param='crps_ss',title=measurement.upper()+'_SS',cmap_label='',save=True,path=save_path+'/'+'single_leading_time_figures/calibration/skill_score_with_calimatology_'+str(climatology_num)+'/50_station/calibration_50_station_single_leading_time_'+str(lead_time+1))

                    acc_ss_pr=1- (acc_pr/(lead_time+1))/climat
                    acc_ss_pr_zg=1- (acc_pr_zg/(lead_time+1))/climat
                    acc_ss_cali=1- (acc_cali/(lead_time+1))/climat
                    acc_ss_bi=1- (acc_bi/(lead_time+1))/climat    

                    print((acc_bi/(lead_time+1)).mean(),BI.mean())

                    dpt.draw_aus_figure(acc_ss_bi,lat,lon,param='acc_crps_ss',title=measurement.upper()+'_SS',cmap_label='',save=True,path=save_path+'/'+'accumulated_leading_time_figures/bi/skill_score_with_calimatology_'+str(climatology_num)+'/whole/bi_whole_accumulated_leading_time_'+str(lead_time+1))
                    dpt.draw_aus_figure(acc_ss_pr_zg,lat,lon,param='acc_crps_ss',title=measurement.upper()+'_SS',cmap_label='',save=True,path=save_path+'/'+'accumulated_leading_time_figures/vdsrd2/skill_score_with_calimatology_'+str(climatology_num)+'/whole/vdsrd2_whole_accumulated_leading_time_'+str(lead_time+1))
                    dpt.draw_aus_figure(acc_ss_pr,lat,lon,param='acc_crps_ss',title=measurement.upper()+'_SS',cmap_label='',save=True,path=save_path+'/'+'accumulated_leading_time_figures/vdsrd/skill_score_with_calimatology_'+str(climatology_num)+'/whole/vdsrd_whole_accumulated_leading_time_'+str(lead_time+1))
                    dpt.draw_aus_figure(acc_ss_cali,lat,lon,param='acc_crps_ss',title=measurement.upper()+'_SS',cmap_label='',save=True,path=save_path+'/'+'accumulated_leading_time_figures/calibration/skill_score_with_calimatology_'+str(climatology_num)+'/whole/calibration_whole_accumulated_leading_time_'+str(lead_time+1))
                    dpt.draw_station_figure(acc_ss_bi,param='acc_crps_ss',title=measurement.upper()+'_SS',cmap_label='',save=True,path=save_path+'/'+'accumulated_leading_time_figures/bi/skill_score_with_calimatology_'+str(climatology_num)+'/50_station/bi_50_station_accumulated_leading_time_'+str(lead_time+1))
                    dpt.draw_station_figure(acc_ss_pr_zg,param='acc_crps_ss',title=measurement.upper()+'_SS',cmap_label='',save=True,path=save_path+'/'+'accumulated_leading_time_figures/vdsrd2/skill_score_with_calimatology_'+str(climatology_num)+'/50_station/vdsrd2_50_station_accumulated_leading_time_'+str(lead_time+1))
                    dpt.draw_station_figure(acc_ss_pr,param='acc_crps_ss',title=measurement.upper()+'_SS',cmap_label='',save=True,path=save_path+'/'+'accumulated_leading_time_figures/vdsrd/skill_score_with_calimatology_'+str(climatology_num)+'/50_station/vdsrd_50_station_accumulated_leading_time_'+str(lead_time+1))
                    dpt.draw_station_figure(acc_ss_cali,param='acc_crps_ss',title=measurement.upper()+'_SS',cmap_label='',save=True,path=save_path+'/'+'accumulated_leading_time_figures/calibration/skill_score_with_calimatology_'+str(climatology_num)+'/50_station/calibration_50_station_accumulated_leading_time_'+str(lead_time+1))







            ####################### #################
#             my_pr=np.load("../save/"+measurement+"/vdsrd/"+str(year_generate)+"/lead_time_"+str(lead_time)+".npy").mean(0)[~np.isinf(land)]
#             my_pr_zg=np.load("../save/"+measurement+"/vdsrd2/"+str(year_generate)+"/lead_time_"+str(lead_time)+".npy").mean(0)[~np.isinf(land)]

#             BI=np.load("../save/"+measurement+"/bi_217/"+str(year_generate)+"/lead_time_"+str(lead_time)+".npy").mean(0)[~np.isinf(land)]
#             climat=load_climatology_data(lead_time,measurement,year_generate,climatology_num).mean(0)[~np.isinf(land)]
#             cali=np.load("../save/"+measurement+"/calibration/"+str(year_generate)+"/lead_time_"+str(lead_time)+".npy").mean(0)[~np.isinf(land)]


#             ss_my_pr= 1-my_pr/climat
#             ss_my_pr_zg=1-my_pr_zg/climat
#             ss_cali=1-cali/climat
#             ss_bi=1-BI/climat





#             save_csv_mean(my_pr,my_pr_zg,cali, BI,climat, ss_my_pr,ss_my_pr_zg,ss_cali,ss_bi, lead=lead_time,file_dir=file_path ,climatology_num=climatology_num)
#             save_csv_median(my_pr,my_pr_zg,cali, BI,climat, ss_my_pr,ss_my_pr_zg,ss_cali,ss_bi, lead=lead_time,file_dir=file_path,climatology_num=climatology_num)

#             my_pr_against_climatology=[]
#             my_pr_zg_against_climatology=[]
#             cali_against_climatology=[]
#             BI_against_climatology=[]
            
#             for q in interval_band:
#                 my_pr_against_climatology.append(np.percentile( ss_my_pr,q))
#                 my_pr_zg_against_climatology.append(np.percentile(ss_my_pr_zg ,q))
#                 cali_against_climatology.append(np.percentile( ss_cali,q))
#                 BI_against_climatology.append(np.percentile(ss_bi ,q))

#             data_my_pr_against_climatology.append(my_pr_against_climatology)
#             data_my_pr_zg_against_climatology.append(my_pr_zg_against_climatology)
#             data_cali_against_climatology.append(cali_against_climatology)
#             data_BI_against_climatology.append(BI_against_climatology)

#         dpt.mkdir(save_path+'/interval_band/')
#         draw_plot(data_my_pr_against_climatology,save_path+'/interval_band/DL_pr_only_against_climatology_'+str(year_generate)+'_'+option+'_climatology_'+str(climatology_num),param=measurement.upper()+'_SS') 
#         draw_plot(data_my_pr_zg_against_climatology,save_path+'/interval_band/DL_pr_zg_against_climatology_'+str(year_generate)+'_'+option+'_climatology_'+str(climatology_num),param=measurement.upper()+'_SS') 
#         draw_plot(data_cali_against_climatology,save_path+'/interval_band/calibration_against_climatology_'+str(year_generate)+'_'+option+'_climatology_'+str(climatology_num),param=measurement.upper()+'_SS') 
#         draw_plot(data_BI_against_climatology,save_path+'/interval_band/BI_against_climatology_'+str(year_generate)+'_'+option+'_climatology_'+str(climatology_num),param=measurement.upper()+'_SS') 

        
        
        
    else:
        
        save_path='../save/results/'+measurement+'/'+str(year_generate)

#         file_path=save_path+'/table_'+option+'_' +str(year_generate)+'_climatology_' +str(climatology_num)
 
#         save_csv_mean(file_dir=file_path,first_init=1)
#         save_csv_median(file_dir=file_path,first_init=1)
        
#         if option=='50':
#             from util.constant_param import station_50_index_for_size_of_hr_sr as station_dict
#         if option=='214':
#             from util.constant_param import station_214_index_for_size_of_hr_sr as station_dict
            
#         land=np.load("../save/mae/calibration/1997/lead_time_"+str(216)+".npy").mean(0)

#         station_index=np.zeros((land.shape))
#         for i in station_dict.keys():
#             station_index[station_dict[i]]=1
            



#         data_my_pr_against_climatology=[]
#         data_my_pr_zg_against_climatology=[]
#         data_cali_against_climatology=[]
#         data_BI_against_climatology=[]

        
#         mean_my=[]



#         for lead_time in range(total_leading):
#             my_pr=np.load("../save/"+measurement+"/vdsrd/"+str(year_generate)+"/lead_time_"+str(lead_time)+".npy").mean(0)
#             my_pr_zg=np.load("../save/"+measurement+"/vdsrd2/"+str(year_generate)+"/lead_time_"+str(lead_time)+".npy").mean(0)

#             BI=np.load("../save/"+measurement+"/bi_217/"+str(year_generate)+"/lead_time_"+str(lead_time)+".npy").mean(0)
#             climat=load_climatology_data(lead_time,measurement,year_generate,climatology_num).mean(0)
#             cali=np.load("../save/"+measurement+"/calibration/"+str(year_generate)+"/lead_time_"+str(lead_time)+".npy").mean(0)         
# #                 draw_overview_figure(my_pr)



#             my_pr=np.load("../save/"+measurement+"/vdsrd/"+str(year_generate)+"/lead_time_"+str(lead_time)+".npy").mean(0)[station_index==1]
#             my_pr_zg=np.load("../save/"+measurement+"/vdsrd2/"+str(year_generate)+"/lead_time_"+str(lead_time)+".npy").mean(0)[station_index==1]
#             BI=np.load("../save/"+measurement+"/bi_217/"+str(year_generate)+"/lead_time_"+str(lead_time)+".npy").mean(0)[station_index==1]
#             climat=load_climatology_data(lead_time,measurement,year_generate,climatology_num).mean(0)[station_index==1]
#             cali=np.load("../save/"+measurement+"/calibration/"+str(year_generate)+"/lead_time_"+str(lead_time)+".npy").mean(0)[station_index==1]


#             ss_my_pr= 1-my_pr/climat
#             ss_my_pr_zg=1-my_pr_zg/climat
#             ss_cali=1-cali/climat
#             ss_bi=1-BI/climat


#             save_csv_mean(my_pr,my_pr_zg,cali, BI,climat, ss_my_pr,ss_my_pr_zg,ss_cali,ss_bi, lead=lead_time,file_dir=file_path ,climatology_num=climatology_num)
#             save_csv_median(my_pr,my_pr_zg,cali, BI,climat, ss_my_pr,ss_my_pr_zg,ss_cali,ss_bi, lead=lead_time,file_dir=file_path,climatology_num=climatology_num)
            
#             my_pr_against_climatology=[]
#             my_pr_zg_against_climatology=[]
#             cali_against_climatology=[]
#             BI_against_climatology=[]
#             for q in interval_band:
#                 my_pr_against_climatology.append(np.percentile( ss_my_pr,q))
#                 my_pr_zg_against_climatology.append(np.percentile(ss_my_pr_zg ,q))
#                 cali_against_climatology.append(np.percentile( ss_cali,q))
#                 BI_against_climatology.append(np.percentile(ss_bi ,q))

#             data_my_pr_against_climatology.append(my_pr_against_climatology)
#             data_my_pr_zg_against_climatology.append(my_pr_zg_against_climatology)
#             data_cali_against_climatology.append(cali_against_climatology)
#             data_BI_against_climatology.append(BI_against_climatology)

#         dpt.mkdir(save_path+'/interval_band/')
#         draw_plot(data_my_pr_against_climatology,save_path+'/interval_band/DL_pr_only_against_climatology_'+str(year_generate)+'_'+option+'_climatology_'+str(climatology_num),param=measurement.upper()+'_SS') 
#         draw_plot(data_my_pr_zg_against_climatology,save_path+'/interval_band/DL_pr_zg_against_climatology_'+str(year_generate)+'_'+option+'_climatology_'+str(climatology_num),param=measurement.upper()+'_SS') 
#         draw_plot(data_cali_against_climatology,save_path+'/interval_band/calibration_against_climatology_'+str(year_generate)+'_'+option+'_climatology_'+str(climatology_num),param=measurement.upper()+'_SS') 
#         draw_plot(data_BI_against_climatology,save_path+'/interval_band/BI_against_climatology_'+str(year_generate)+'_'+option+'_climatology_'+str(climatology_num),param=measurement.upper()+'_SS') 


# if __name__=='__main__':

#     for op in ['whole','50']:
#         for window in [11]:
#             for y in [1997]:
#                 for m in ['crps']:#['mae','rmse','crps']
#                     draw_confidence_figure(m,window,op,y)

import time
if __name__=='__main__':

    for op in ['whole','50']:
        for window in [1,11]:
            for y in [1997,2010,2012]:
                for m in ['mae','rmse','crps','bcsd']:#
                    t=time.time()
                    
                    draw_confidence_figure(m,window,op,y)
                    print(op+str(window)+str(y)+m)
                    print(time.time()-t)

        
```
